#### After checking all the data from the World Happiness Report of the years 2015-2020, we can see that there is a difference in the number of rows per year, which means that not all countries have a happiness report in a certain year. Thus, we only select the countries that have a record starting from year 2015 to 2020. Since the year 2020 has the least number of countries, we will use it as a base and cross check if the countries in the 2020 dataset is also in the other datasets.

For this process we used MS Excel and found out that all the countries in the dataset for year 2020 is present in other datasets.

## Creating the Final Dataset

```python
whrfinalCountries=pd.DataFrame(whr2020['Country'])
whrfinalCountries['Region']=whr2020['Regional indicator']
whrfinalCountries
```

```python
whrfinalCountries.isnull().sum()
```

```python
whr2015final = whr2015[whr2015.Country.isin(whrfinalCountries.Country)]
del whr2015final['Dystopia Residual']
del whr2015final['Family']
del whr2015final['Standard Error']
del whr2015final['Region']
whr2015final = whr2015final.rename(columns={'Happiness Rank':'Happiness_Rank','Happiness Score':'Happiness_Score',
                                            'Economy (GDP per Capita)':'GDP_hscore','Health (Life Expectancy)':'LifeExp_hscore',
                                           'Freedom':'Freedom_hscore', 'Trust (Government Corruption)':'Corruption_hscore',
                                            'Generosity':'Generosity_hscore'}
                                  ,errors="raise")
whr2015final=pd.merge(whr2015final,whrfinalCountries, on='Country', how='left') 
whr2015final
```

```python
whr2016final = whr2016[whr2016.Country.isin(whrfinalCountries.Country)]
del whr2016final['Dystopia Residual']
del whr2016final['Family']
del whr2016final['Lower Confidence Interval']
del whr2016final['Upper Confidence Interval']
del whr2016final['Region']
whr2016final = whr2016final.rename(columns={'Happiness Rank':'Happiness_Rank','Happiness Score':'Happiness_Score',
                                           'Economy (GDP per Capita)':'GDP_hscore','Health (Life Expectancy)':'LifeExp_hscore',
                                           'Freedom':'Freedom_hscore', 'Trust (Government Corruption)':'Corruption_hscore','Generosity':'Generosity_hscore'}
                                   ,errors="raise")
whr2016final=pd.merge(whr2016final,whrfinalCountries, on='Country', how='left')
whr2016final
```

```python
whr2017final = whr2017[whr2017.Country.isin(whrfinalCountries.Country)]
del whr2017final['Dystopia.Residual']
del whr2017final['Family']
del whr2017final['Whisker.high']
del whr2017final['Whisker.low']
whr2017final = whr2017final.rename(columns={'Happiness.Rank':'Happiness_Rank','Happiness.Score':'Happiness_Score',
                                           'Economy..GDP.per.Capita.':'GDP_hscore','Health..Life.Expectancy.':'LifeExp_hscore',
                                           'Freedom':'Freedom_hscore', 'Trust..Government.Corruption.':'Corruption_hscore','Generosity':'Generosity_hscore'}
                                   ,errors="raise")
whr2017final=pd.merge(whr2017final,whrfinalCountries, on='Country', how='left')
whr2017final
```

```python
whr2018final = whr2018[whr2018.Country.isin(whrfinalCountries.Country)]
del whr2018final['Social support']
whr2018final = whr2018final.rename(columns={'Overall rank':'Happiness_Rank','Score':'Happiness_Score',
                                           'GDP per capita':'GDP_hscore','Healthy life expectancy':'LifeExp_hscore',
                                           'Freedom to make life choices':'Freedom_hscore', 'Perceptions of corruption':'Corruption_hscore',
                                            'Generosity':'Generosity_hscore'}
                                   ,errors="raise")
whr2018final=pd.merge(whr2018final,whrfinalCountries, on='Country', how='left')
whr2018final
```

```python
whr2019final = whr2019[whr2019.Country.isin(whrfinalCountries.Country)]
del whr2019final['Social support']
whr2019final = whr2019final.rename(columns={'Overall rank':'Happiness_Rank','Score':'Happiness_Score',
                                           'GDP per capita':'GDP_hscore','Healthy life expectancy':'LifeExp_hscore',
                                           'Freedom to make life choices':'Freedom_hscore', 'Perceptions of corruption':'Corruption_hscore',
                                            'Generosity':'Generosity_hscore'}
                                   ,errors="raise")
whr2019final=pd.merge(whr2019final,whrfinalCountries, on='Country', how='left')
whr2019final
```

```python
whr2020final = whr2020[whr2020.Country.isin(whrfinalCountries.Country)]
del whr2020final['Standard error of ladder score']
del whr2020final['upperwhisker']
del whr2020final['lowerwhisker']
del whr2020final['Logged GDP per capita']
del whr2020final['Social support']
del whr2020final['Healthy life expectancy']
del whr2020final['Freedom to make life choices']
del whr2020final['Generosity']
del whr2020final['Perceptions of corruption']
del whr2020final['Ladder score in Dystopia']
del whr2020final['Dystopia + residual']
del whr2020final['Explained by: Social support']

whr2020final = whr2020final.rename(columns={'Rank':'Happiness_Rank','Regional indicator':'Region','Ladder score':'Happiness_Score',
                                           'Explained by: Log GDP per capita':'GDP_hscore','Explained by: Healthy life expectancy':'LifeExp_hscore',
                                           'Explained by: Freedom to make life choices':'Freedom_hscore', 
                                            'Explained by: Perceptions of corruption':'Corruption_hscore',
                                            'Explained by: Generosity':'Generosity_hscore'}
                                   ,errors="raise")
whr2020final
```

```python
whrfinal=whr2020final
whrfinal=whrfinal.append(whr2016final,ignore_index = True)
whrfinal=whrfinal.append(whr2017final,ignore_index = True)
whrfinal=whrfinal.append(whr2018final,ignore_index = True)
whrfinal=whrfinal.append(whr2019final,ignore_index = True)
whrfinal=whrfinal.append(whr2015final,ignore_index = True)
whrfinal

```
